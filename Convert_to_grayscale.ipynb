````python
def convert2Grayscale(folderPath: str, folder: str):
  import os
  for label in os.listdir(folderPath):
    if os.path.isdir(folderPath):
      labelPath = os.path.join(folderPath, label)
      newLabelPath = os.path.join(newPath, folder, label)
      os.makedirs(newLabelPath)
      if os.path.isdir(labelPath):
        for imgFile in os.listdir(labelPath):
          try:
            imgPath = os.path.join(labelPath, imgFile)
            image = cv2.imread(imgPath)
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            newImgPath = os.path.join(newLabelPath, imgFile)
            cv2.imwrite(os.path.join(newImgPath), gray)
            print("Converted : ", newImgPath)
          except:
            print("It is not an image : ", imgPath)
      else:
        print("Not a directory : ", labelPath)
    else:
      print("Not a directory : ", folderPath)
````

Dont Forget to change your Dataset path and new converted path before running ths

````python
import sys
import os
import shutil
import cv2
path = 'Dataset'
newPath = 'Grayscale' 
try:
  os.mkdir(newPath)
except:
  print("Directory exist")
  x = input("Do you want to delete all file and directory inside exist path? : ")
  if x.lower() == "y":
    shutil.rmtree(newPath, ignore_errors=False, onerror=None)
    os.mkdir(newPath)
  else :
    sys.exit()

for folder in os.listdir(path):
  folderPath = os.path.join(path, folder)
  if os.path.isdir(folderPath):  
    if folder.lower() == "testing" or folder.lower() == "training":
      convert2Grayscale(folderPath, folder)
  else:  
    print("\n It is not a directory" )
````
